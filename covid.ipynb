```python
import pandas as pd
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv('COVID.csv')
```

```python
df
```

```python
df.shape
```

```python
train, test = train_test_split(df, test_size=0.3)
```

```python
print(train.shape)
```

```python
print(test.shape)
```

```python
train_x = train[['Gender','Age', 'Nocomorbidity','Cardiovascular','Endocrine', 'tumor', 'Respiratory', 'Digestive', 'Renal', 'Liver', 'Fever', 'Cough', 'Chest tightness', 'Fatigue', 'Diarrhea', 'RNA']]
train_y = train.COVID
```

```python
test_x = train[['Gender','Age', 'Nocomorbidity','Cardiovascular','Endocrine', 'tumor', 'Respiratory', 'Digestive', 'Renal', 'Liver', 'Fever', 'Cough', 'Chest tightness', 'Fatigue', 'Diarrhea', 'RNA']]
test_y = train.COVID
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
knn = KNeighborsClassifier(n_neighbors=16)
knn.fit(train_x, train_y)
```

```python
prediction = knn.predict(test_x)
```

```python
print("Accuracy:",metrics.accuracy_score(prediction, test_y))
```

```python
pickle.dump(knn,open('covid.pkl','wb'))
```
